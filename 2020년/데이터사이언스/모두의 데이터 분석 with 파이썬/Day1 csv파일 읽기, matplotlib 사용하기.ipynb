```python
#csv(comma Separated Values) 데이터 불러오기

#1. CSV 라이브러리 불러오기
import csv

#2. csv 파일 열기
f = open("seoul_temp.csv")

#3. 데이터를 콤마로 구분하여 저장하기
data = csv.reader(f)

#4. for 반복문으로 한 줄씩 꺼내기.
for row in data:
    print(row)
```
