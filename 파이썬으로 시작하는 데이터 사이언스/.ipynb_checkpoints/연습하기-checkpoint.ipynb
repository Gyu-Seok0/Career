# Folium 이용해보기
강동구 명일동에 있는 음식점 데이터를 분석해보자.

```python
import folium
```

```python
df_east_food = df[(df["법정동명"].str.contains("명일")) & (df["상권업종대분류명"] == "음식")]
df_east_food["경도"].mean()
df_east_food["위도"].mean()
```

```python
colnames = df_east_food.columns.tolist()

for i in colnames:
    print(i, ":", df_east_food.loc[53,i])
```

```python
map = folium.Map(location = [df_east_food["위도"].mean(),df_east_food["경도"].mean()],zoom_start = 15)
for n in df_east_food.index:
    name = df_east_food.loc[n,"상호명"]
    address = df_east_food.loc[n,"도로명주소"]
    popup = f"{name}--{address}"
    location = [df_east_food.loc[n,"위도"], df_east_food.loc[n,"경도"]]
    folium.Marker(
        location = location,
        popup = popup,
    ).add_to(map)
map

```
