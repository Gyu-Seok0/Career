```python
import csv
import matplotlib.pyplot as plt

f = open("population.csv")
data = csv.reader(f)
next(data)

mytown=[0]*101
for row in data:
    if '천호' in row[0]:
        for i in range(3,104):  # 101개 여야함.. 맞네!
            mytown[i-3] += int(row[i])
            
#저장된 데이터를 꺾은선 그래프로 그리기.
#plt.figure(dpr= 300)
plt.style.use('ggplot')
plt.title('population of chen-ho-dong')
plt.plot(mytown)
plt.xlabel("age")
plt.ylabel("nums")
plt.savefig('천호동.jpeg')
plt.show()
```
