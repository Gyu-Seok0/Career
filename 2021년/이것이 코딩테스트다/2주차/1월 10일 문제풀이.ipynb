# 나올 수 있는 가장 큰 수를 생각해보자
DFS로 문제를 풀이해보자

숫자는 고정이고, 연산을 우리가 다양하게 적용이 가능하다.

다 해봐야한다 --> 부르트 포스


```python
N = int(input())
num = list(map(int,input().split()))
op = list(map(int,input().split()))

min_x, max_x = 1e11,-1e11
def dfs(result, cnt):
    global min_x, max_x
    
    if cnt == N-1:  #종료구문
        min_x = min(min_x, result)
        max_x = max(max_x, result)
        return

    for i in range(4): #연산구문을 리커시브로 돌리자
        if op[i] > 0:
            op[i] -= 1
            if i == 0:
                dfs(result + num[cnt+1], cnt+1)
            elif i == 1:
                dfs(result - num[cnt+1], cnt+1)
            elif i == 2:
                dfs(result * num[cnt+1], cnt+1)
            elif i == 3: 
                dfs(int(result / num[cnt+1]), cnt+1)
            op[i] += 1 #다시 추가해줘야 반복적인 연산이 가능하다.
            

dfs(num[0],0)
print(max_x)
print(min_x)
```

1) op[i] += 1 을 마지막에 해줄 것 --> 그래야 백 트래킹이 가능하다.

2) min, max함수의 적절한 활용

3) global 변수를 선언하기 위해서는 함수 안에서 선언

4) 1e-10 과 같은 변수 선언 가능

```python
from collections import deque

N = int(input())
graph = []
s = []
t = []
for i in range(N):
    graph.append(input().split())
    
    for j in range(N):
        if graph[i][j] == "S": #이렇게 하면 시간복잡도를 훨씬 줄일수 있지!
            s.append((i,j))
        elif graph[i][j] == "T":
            t.append((i,j))


queue = deque(s)

def solution():
    obstacle = []
    cnt = 1
    while queue:
        x,y = queue.popleft()
        #row 확인
        for teacher in t:
            if x == teacher[0]: # 같다면
                if abs(y - teacher[1]) == 1:
                    return "NO"
                else:
                    for i in range(min(teacher[1],y)+1, max(teacher[1],y)):

                        obstacle.append((x,i,cnt))
                    cnt += 1
            if y == teacher[1]:
                if abs(x - teacher[0]) == 1:
                    return "NO"
                else:
                    for j in range(min(teacher[0],x)+1, max(teacher[0],x)):
                        obstacle.append((j,y,cnt))
                    cnt += 1
                    
    count = cnt-1
    for i in range(len(obstacle)):
        nx,ny,target = obstacle[i]
        for j in range(i,len(obstacle)):
            dx, dy, dtarget = obstacle[j]
            if target == dtarget:
                continue
            elif nx == dx and ny == dy:
                count -= 1
                break

    if count == 3:
        return "NO"
    else:
        return "YES"
                               
#col 확인
print(solution())
```

### 동빈's 풀이

```python
from itertools import combinations

n = int(input())
board = []
teachers = []
spaces = []

for i in range(n):
    board.append(input().split())
    for j in range(n):
        if board[i][j] == "T":
            teachers.append((i,j))
        elif board[i][j] == "X":
            spaces.append((i,j))
            
def watch(x,y,direction):
    #왼쪽
    if direction == 0:
        while y >= 0:
            if board[x][y] == "S":
                return True
            if board[x][y] == "O":
                return False
            y -= 1
    
    # 오른쪽
    if direction == 1:
        while y < n:
            if board[x][y] == "S":
                return True
            if board[x][y] == "O":
                return False
            y += 1 
    #위       
    if direction == 2:
        while x >= 0:
            if board[x][y] == "S":
                return True
            if board[x][y] == "O":
                return False
            x -= 1
    #아래
    if direction == 3:
        while x < n:
            if board[x][y] == "S":
                return True
            if board[x][y] == "O":
                return False
            x += 1
    return False

# 장애물 설치 이후, 학생이 발견되는지 check
def process():
    for x,y in teachers:
        for i in range(4):
            if watch(x,y,i):
                return True
    return False

find = False

for data in combinations(spaces,3):
    for x,y in data:
        board[x][y] = 'O'
    if not process():
        find = True
        break
    for x,y in data:
        board[x][y] = "X"

if find:
    print("YES")
else:
    print("NO")
    
    
    
    
    
    
    
    
```
