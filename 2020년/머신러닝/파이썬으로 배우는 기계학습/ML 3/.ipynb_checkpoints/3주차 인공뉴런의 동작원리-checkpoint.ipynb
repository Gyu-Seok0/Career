# AND 게이트 뉴론 구현하기

```python
import numpy as np
import matplotlib.pyplot as plt

def AND(x1,x2):
    x = np.array([1,x1,x2])
    w = np.array([-0.7,0.5,0.5])  # 리스트 형태로 만드는 것을 잊지말기
    return int(np.dot(w,x) > 0)
```

```python
#AND게이트 뉴론 테스트
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
#AND게이트 뉴론 시각화
# -0.7+ 0.5*X1+ 0.5X2 = 0  --> x2 = -x1+1.4
x1 = [a for a in range(5)]     
x2 = [-b+1.4 for b in range(5)]
a = [0,0,1,1]
b = [0,1,0,1]

plt.plot(x1,x2, '-sr',a,b,'ob')  # (1,1)만 해당 방정식을 넘어서므로, AND GATE는 (1,1)에서만 1값을 가진다는 것을 알 수 있다!
plt.show()  # 이와 같은 과정을 Classification 이라고 한다. 

```

# 활성화함수

```python
# 섭씨를 화씨로 변화시키는 인공뉴런

def activate(z):
    if z<32:
        z = 32
    return z

def C2F(C):
    F = 9/5.0*C+32
    return activate(F)

#test_c = np.array([-20,-10,0,36.5,40,50,100])
#test_f = C2F(test_c)
test_c = [-20,-10,0,36.5,40,50,100]
test_f = [C2F(c) for c in test_c]

# 하나의 인공뉴런을 두 개의 함수로 구현된 것을 볼 수 있다.
```

```python
test_f
```

```python
x = np.arange(-100,100,.1)
y = [C2F(c) for c in x]

plt.figure()
plt.plot(x,y)
plt.axis([-20,50,0,150])  #그래프의 축을 조정해서 보여준다. 
plt.xlabel('Celcius')
plt.ylabel('Fahrenheit')
plt.show()
```

### 시그모이드(sigmoid)

```python
def sigmoid(x):
    return 1/ (1+np.exp(-x))

x = np.arange(-10.0, 10.0, 0.1)
y = sigmoid(x)
plt.plot(x,y)
plt.axvline(0, color = "black")
plt.xlabel('x',fontsize=16)
plt.ylabel('h(x)', fontsize = 16)
plt.ylim(-0.1,1.1)
plt.yticks([0.0, 0.5, 1.0])
plt.grid(axis = 'y')
plt.show()
```
