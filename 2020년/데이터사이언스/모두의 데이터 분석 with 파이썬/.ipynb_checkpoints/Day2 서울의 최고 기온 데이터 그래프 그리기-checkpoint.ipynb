```python
import csv
import matplotlib.pyplot as plt

f = open('test.csv')
data = csv.reader(f)
next(data)

max_temp = []
for row in data:
    max_temp.append(float(row[-1])) # float조심하기.
```

```python
plt.plot(max_temp)
plt.show()
```

```python
plt.boxplot(max_temp)
plt.show()
```

```python
plt.hist(max_temp,bins=30)
plt.show()
```
