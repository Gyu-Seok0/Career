```python
import urllib.parse
import urllib.request
from bs4 import BeautifulSoup
```

```python
baseurl = "https://search.naver.com/search.naver?where=post&sm=tab_jum&query="
plusurl = input("검색어 입력: ")
pageurl = "&sm=tab_pge&srchby=all&st=sim&where=post&start="

url = baseurl + urllib.parse.quote_plus(plusurl)
data = []
for i in range(1,52,10):
    temp = pageurl + str(i)
    temp2 = url + temp
    html = urllib.request.urlopen(temp2).read()
    soup = BeautifulSoup(html, 'html.parser')
    title = soup.find_all(class_ = "sh_blog_title")
    
    for i in title:
        print(i.attrs['title'])
        print(i.attrs['href'])
        print()
        data.append([i.attrs['title'],i.attrs['href']])



```

```python
#data를 csv파일에 저장하기.
with open('python.csv','w') as file:
    file.write('제목, url\n')
    for i in data:
        file.write("{0},{1}\n".format(i[0],i[1]))
```

```python
import urllib.request
import pandas as pd
from bs4 import BeautifulSoup
print("정치(0)/경제(1)/사회(2)/문화(3)/세계(4)/IT(5)")
subject = input("원하는 기사 숫자로 입력: ")

#url 만들기
baseurl = "https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=10"
pageurl = "#&date=%2000:00:00&page="
url = baseurl+subject+pageurl

#data 저장 및 출력
data = []
for i in range(1,2):   # 원하는 페이지 지정. 예를 들어 5page까지 표현하고 싶으면, range(1,6)으로 표현하면 됩니다.
    temp = url + str(i)
    html = urllib.request.urlopen(temp).read()
    soup = BeautifulSoup(html, 'html.parser')
    title = soup.find_all(class_ = 'cluster_text_headline')
    for j in title:
        temp2 = str(j).split('>')
        name = temp2[1].strip('</a')
        link = j.attrs['href']
        print(name)
        print(link)
        print()
        data.append([name,link])

# 파일로 저장하기.
df = pd.DataFrame(data)
df.columns = ['제목', '링크']                  # columns의 이름을 지정해줍니다. 
df.head()

df.to_csv('sample2.csv', encoding = "cp949")   #현재 디렉토리에 파일이 저장됩니다. 파일명은 sample2가 되겠습니다.
```
