파이썬을 이용해서 네이버 블로그 검색창 제목과 url을 가져왔다.
굉장히 간단하다.

```python
import urllib.parse
import urllib.request
from bs4 import BeautifulSoup
```

```python
'https://search.naver.com/search.naver?sm=tab_hty.top&where=post&query=%EC%9E%90%EB%B0%94'
baseurl = "https://search.naver.com/search.naver?where=post&sm=tab_jum&query="
plusurl = input('검색어를 입력하세요: ')
pageurl = '&sm=tab_pge&srchby=all&st=sim&where=post&start='


url = baseurl + urllib.parse.quote_plus(plusurl)   # plus url이 한글 일때.. encoding 해주는 역할!

for i in range(1,52,10):
    temp = pageurl + str(i)
    temp2 = url+temp
    html = urllib.request.urlopen(temp2).read()
    soup = BeautifulSoup(html, 'html.parser') #모든 html 긁어오기
    title = soup.find_all(class_='sh_blog_title') # class가 sh_blog_title인거 다 긁어오기.
    
    for i in title:    
        print(i.attrs['title'])
        print(i.attrs['href'])
        print()
```

```python
#한동대 사이트
# error 1 page를 바꿀때마다 원래라면 새로운 url이 만들어져야 하는데, 그렇지 아니하다.
# error 2 네이버처럼 dict 형태가 아니라서, 구문을 새롭게 짜야함.
# error 3 각 기사별 link가 규칙성이 존재하지 않는다.. 아니면 내가 못발견하거나.. 보통 1씩 차이나는데, 이건 좀 이상하다.
import urllib.request
from bs4 import BeautifulSoup

url = 'https://www.handong.edu/square/inform/press/?mode=list'

html = urllib.request.urlopen(url).read()
soup = BeautifulSoup(html,'html.parser')
title = soup.find_all(class_='title al')

for i in title:
    temp = str(i).split('>')
    print(temp[2])
    print()


```

```python
#히즈넷
#히즈넷은 기본적으로 로그인이 되야 실행이 가능한 것 같다. --> 아직 잘 모르겠누
import urllib.request
from bs4 import BeautifulSoup

url = "https://hisnet.handong.edu"
html = urllib.request.urlopen(url).read()
soup = BeautifulSoup(html,'html.parser')
title = soup.find_all(href = "javascript:alert('로그인하여주십시오.')")

for i in title:
    print(i)
    break
```

```python
#네이버 기사 헤드라인을 읽어오자.
import urllib.request
from bs4 import BeautifulSoup

# input
print("정치(0)/경제(1)/사회(2)/문화(3)/세계(4)/IT(5)")
subject = input("원하는 기사 숫자로 입력: ")

#url 만들기
url = "https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=10"
url += subject
html = urllib.request.urlopen(url).read()
soup = BeautifulSoup(html,'html.parser')
title = soup.find_all(class_ = 'cluster_text_headline')

# output
for i in title:
    temp = str(i).split('>')
    name = temp[1].strip('</a')
    index = temp[0].find('https')
    print(name)
    print(temp[0][index:])
    print()

```
