```python
def play(row,col,cnt):
    if chess[row][col] != 0:
        return False #체스를 둘 수 없는 경우
    else:
        #가로/세로에 놓기
        for i in range(N):
            chess[row][i] = 'X' # 0 ~ N-1
            chess[i][col] = 'X'

        #대각선에 놓기 
        dx = [-1,-1,1,1] # 좌상,우상,좌하,우하
        dy = [-1,1,-1,1]
        for i in range(4):
            tx = row + dx[i]
            ty = col + dy[i]
            while True:
                if tx < 0 or tx >= N or ty < 0 or ty >= N:
                    break
                else:
                    chess[tx][ty] = 'X'
                    tx += dx[i]
                    ty += dy[i]
        #자기자신 놓기
        chess[row][col] = 1
        
        print("row = {}, col = {}, cnt = {}".format(row,col,cnt))
        for i in range(N):
            for j in range(N):
                print(chess[i][j], end="")
            print()
        return True #체스를 뒀고, 1과 -1 Mapping완료
```

```python
import copy
chess = [[0]*(N) for _ in range(N)]
before_chess = copy.deepcopy(chess)
#row = 0 #row
#col = 0 #col
cnt = 0 #퀸 개수
case = 0 #총 경우의수
def dfs(row,col,cnt,chess):
    global case,before_chess
    if cnt == N:
        case += 1
        return True #N개를 무사히 놨다.
    
    if row == N-1 and col == N-1:    
        return False #N개를 다 놓지 못하고 끝에 도달했다.
    
    #cnt 갱신
    if play(row,col,cnt):
        before_chess = copy.deepcopy(chess)
        n_cnt = cnt + 1
    else:
        n_cnt = cnt
    
    #row/col 갱신
    if col == N-1:
        n_col = 0
        n_row = row + 1
    else:
        n_col = col + 1
        n_row = row

    dfs(n_row,n_col,n_cnt,chess) #다음 퀸을 놓는다
    dfs(n_row,n_col,cnt,before_chess) #이전으로 돌아간다.
        
       
dfs(0,0,cnt,chess)
```

# 세번째 전략: 시간초과

```python
# 퀸이 있는지 없는지 확인
N = int(input())

chess = [[0]*N for _ in range(N)]
case = 0

def check(row,col):
    # 가로/세로
    idx = 0
    while idx < N:
        if chess[row][idx] != 0 or chess[idx][col] != 0:
            return False
        idx += 1
    
    # 대각선: 우하/좌하/우상/좌상
    dx = [1,1,-1,-1]
    dy = [1,-1,1,-1]
    for i in range(4):
        tx = row + dx[i] 
        ty = col + dy[i]
        while True:
            if tx < 0 or tx >= N or ty < 0 or ty >= N:
                break
            elif chess[tx][ty] != 0:
                return False
            tx += dx[i]
            ty += dy[i]
    return True


def dfs(row,col,cnt):
    global case
    # baseline
    if cnt == N:
        return True #N개가 다 채워짐
    if row >= N:
        return False #N개가 채워지기 전에, 끝에 도달함
    
    #recursive
    flag = False
    if check(row,col):
        chess[row][col] = 1
        flag = True
        
        print("row = {}, col = {}, cnt = {}, case = {}".format(row,col,cnt+1,case))
        for i in range(N):
            for j in range(N):
                print(chess[i][j], end = " ")
            print()
        
    n_row, n_col, n_cnt = row,col+1, cnt + flag
    if n_col >= N:
        n_row += 1
        n_col = 0
    
    if dfs(n_row,n_col,n_cnt):
        case += 1
    
    if flag:
        chess[row][col] = 0
        dfs(n_row,n_col,cnt)
    
dfs(0,0,0)
```

```python
#https://codepractice.tistory.com/82
#https://geonlee.tistory.com/40
def nqeen(sol,n):
    global count
    
    if len(sol) == n:
        count += 1
        print(count,sol)
        return 0
    candidate = list(range(n)) #후보자 만들기
    for i in range(len(sol)):
        if sol[i] in candidate: #같은 열에 있다면 후보에서 제외
            candidate.remove(sol[i])
        distance = len(sol) - i
        if sol[i] + distance in candidate: #같은 대각선(+) 상에 있는가
            candidate.remove(sol[i] + distance)
        if sol[i] - distance in candidate:
            candidate.remove(sol[i] - distance)
    if candidate != []:
        for i in candidate:
            sol.append(i)
            nqeen(sol,n)
            sol.pop()
    else:
        return 0
count = 0
num = int(input())
for i in range(num):
    nqeen([i],num)
print(count)
```
