```python
num = []
for i in range(7):
    num.append(int(input()))

# 제일큰 수 찾기
big = num[0]
second = num[0]

# 제일
flag = True # num[0] = big

for i in range(1,7):
    if big < num[i]:
        flag = False
        second = big
        big = num[i]
        
if flag == True:
    second = num[1] # init
        
for i in range(7):
    if second < num[i] and num[i] < big:
        second = num[i]
print(big, second)
```

```python
# 버블소트
# 자기옆에 있는거랑 비교하기.
temp = [1,5,2,8,3,9,4]

for i in range(len(temp)):
    for j in range(i,len(temp)-1):
        if temp[j] > temp[j+1]:
            temp[j], temp[j+1] = temp[j+1], temp[j]
print(temp)
```
