### 알고리즘 설명

1) 시작노드 선정

2) 최단거리 테이블 초기화

3) 방문하지 않은 노드 중, 최단거리에 있는 노드 선택

4) 해당 노드를 거쳐, 다른 노드로 가는 비용을 계산하고 최단거리 테이블 갱신

5) 위 과정을 3~4번 반복

```python
INF = 1e9
# 노드, edge 수
n,m = map(int, input().split())

# step1,2) 시작노드 선정 & 최단거리 테이블 초기화 (DP방식이구나)
start = int(input())
graph = [[] for _ in range(n+1)] #input graph
visited = [False] * (n+1) #방문체크
distance = [INF] * (n+1) #거리체크

for _ in range(m):
    a,b,c = map(int , input().split())
    graph[a].appned((b,c))
    
# step3) 가장 짧은 노드 반환
def get_smallest_node():
    min_value = INF
    index = 0
    for i in range(1,n+1):
        if distance[i] < min_value and not visited[i]:
            min_value = distance[i]
            index = i
    return index

#step4) 비용계산
def dijstra(start):
    distance[start] = 0
    visited[start] = True
    for j in graph[start]:
        distance[j[0]] = j[1]
        
    for i in range(n-1):
        now = get_smallest_node()
        visited[now] = True
        for j in graph[now]: #이 부분에서 갱신이 이뤄짐.
            cost = distance[now] + j[1]
            if cost < distance[j[0]]:
                distance[j[0]] = cost

                
            
dijstra(start)
for i in range(1,n+1):
    if distance[i] == INF:
        print("무한")
    else:
        print(distance[i])
```

```python
def dijkstar(start):
    q = []
    heapq.heappush(q,(0,start))
    distance[start] = 0
    while q:
        dist, now = heapq.heappop(q)
        if distance[now] < dist: # 이미 처리된거지
            continue
        for i in graph[now]:
            cost = dist + i[1]
            if cost < distance[i[0]]:
                distance[i[0]] = cost
                heapq.heappush(q,(cost, i[0]))
                
```

# 플로이드 워셜 알고리즘

모든 노드에서 다른 모든 노드까지의 최단경로를 찾아보자. <br>
D[a][b] = min(D[a][b],D[a][k] + D[k][b])

```python
INF = int(1e9)

#노드 및 edge 개수
n,m = map(int, input().split())

#2차원 리스트
graph = [[INF]*(n+1) for _ in range(n+1)]

#자기자신 -> 자기자신 = 0
for a in range(1,n+1):
    for b in range(1,n+1):
        if a == b:
            graph[a][b] = 0

# 정보 입력받기
for _ in range(m):
    a,b,c = map(int, input().split())
    graph[a][b] = c
    
for k in range(1,n+1):
    for a in range(1,n+1):
        for b in range(1,n+1):
            graph[a][b] = min(graph[a][b],graph[a][k] + graph[k][b])

```
