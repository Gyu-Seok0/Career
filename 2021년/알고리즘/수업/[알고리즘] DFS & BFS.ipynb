# 그래프 탐색
주어진 그래프를 edge를 따라서 node를 탐색하여 원하는 결과물을 추론하는 알고리즘이다. 방법론으로는 아래 두 가지가 존재한다. 


1) DFS (Depth First Search)

깊이를 우선적으로 탐색하는 알고리즘으로, 재귀함수를 통해 구현 가능하다.

2) BFS (Breadth First Search)

너비를 우선적으로 탐색하는 알고리즘으로, 큐를 사용해서 구현이 가능하다.




### DFS

```python
graph = [
    [],    # 0번 노드는 없으므로, 비워둔다.
    [2,3,8], #1번 노드와 연결된 노드들
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]
```

```python
visited = [False]* len(graph)
def dfs(graph, v, visited):
    visited[v] = True # 방문여부 체크하기
    print(v, end = " ")
    for node in graph[v]:
        if not visited[node]:
            dfs(graph, node, visited)
dfs(graph,1,visited)         
```

### BFS

```python
from collections import deque
visited = [False]*len(graph)
def bfs(graph, start, visited):
    queue = deque([start])
    visited[start] = True
    
    while queue:
        v = queue.popleft()
        print(v, end = " ")
        
        for node in graph[v]:
            if not visited[node]:
                queue.append(node)
                visited[node] = True

bfs(graph,1,visited)
```

# 백준 1260번: DFS와 BFS

```python
from collections import deque

N,M,start = map(int , input().split())
graph = [[] for _ in range(N+1)]

for _ in range(M):
    a,b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)

# 문제에서 낮은 번호부터 search하라고 요구    
for i in range(1,N+1):
    graph[i].sort()

def dfs(graph,v,visited):
    visited[v] = True
    print(v, end = " ")
    for node in graph[v]:
        if not visited[node]:
            dfs(graph,node,visited)
            
def bfs(graph, start, visited):
    queue = deque([start])
    visited[start] = True
    while queue:
        v = queue.popleft()
        print(v, end = " ")
        for node in graph[v]:
            if not visited[node]:
                queue.append(node)
                visited[node] = True
                
visited = [False] * (N+1)
dfs(graph,start,visited)
print()
visited = [False] * (N+1)
bfs(graph, start, visited)            
```

# 백준 1525번: 퍼즐

```python
graph = []
for _ in range(3):
    graph.append(list(map(int, input().split())))
```

```python
from collections import deque

dx = [-1,1,0,0]
dy = [0,0,-1,1]

m = 0
queue, distance = deque(), dict()
for i in range(3):
    for j in range(3):
        n = graph[i][j]
        if not n:
            n = 9
        m = m*10 + n # 하나의 숫자로 입력이 들어옴.
queue.append(m)
distance[m] = 0

history = dict()
history[m] = [m]

bfs()
```

```python
def bfs():
    while queue:
        val = queue.popleft()
        if val == 123894765:
            print(distance[val])
            return
        s = str(val)
        k = s.find('9') # 0의 위치
        x,y = k //3, k%3
        
        for i in range(4):
            nx, ny = x + dx[i], y + dy[i]
            
            if 0 <= nx < 3 and 0 <= ny < 3:
                nk = nx * 3 + ny # 0을 새로운 위치로 이동시킴.
                ns = list(s)
                ns[k], ns[nk] = ns[nk], ns[k] # swap
                nval = int(''.join(ns)) 
                if not distance.get(nval):
                    distance[nval] = distance[val] + 1
                    queue.append(nval)
                if not history.get(nval):
                    history[nval] = history[val] + [nval]
    print(len(queue))
    print(-1)
```
