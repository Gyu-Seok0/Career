```python
#이렇게 하면 for문이랑 논리적으로 같은듯
N = int(input())
arr = [0]*(N+1)

def fibo(num):
    if num < 2:
        arr[num] = num
        return arr[num]
    elif arr[num-1] == 0:
        fibo(num-1)
    arr[num] = arr[num-1] + arr[num-2]
    return arr[num]
fibo(N)
```

```python
# 다른풀이
N = int(input())
arr = [0]*(N+1)

def fibo(num):
    # base1
    if num < 2:
        arr[num] = num
        return num
    #base2
    elif arr[num] != 0:
        return arr[num]
    #recur
    arr[num] = fibo(num-1) + fibo(num-2)
    return arr[num]

print(fibo(N))
```

```python
# https://blog.naver.com/moonsoo5522/220706431458
# 미쳤다.. (y,x)를 기억해야겠다..
# 이거는 내가 암기한다.

N = int(input())
mat = [[" "] *(N) for _ in range(N)]

cnt = 0
def solve(y,x,num):
    global cnt
    cnt += 1
    if num == 1:
        mat[y][x] = "*"
        return
    div = int(num/3)
    for i in range(3):
        for j in range(3):
            if i == 1 and j == 1:
                pass
            else:
                solve(y + (i*div),x + (j*div) ,div)            
solve(0,0,N)
for i in range(N):
    for j in range(N):
        print(mat[i][j], end = "")
    print()
```

```python
N = int(input())
# 이중배열
mat = [[" "]*N for _ in range(N)]
def solve(x,y,num):
    #Base
    if num == 1:
        mat[x][y] = "*"
        return
    div = int(num/3)
    for i in range(3):
        for j in range(3):
            if not (i,j) == (1,1):
                solve(x+i*div ,y+j*div ,div) # x는 j가 다 돌때까지 고정한다.
    
solve(0,0,N)
```

```python
for i in range(N):
    for j in range(N):
        print(mat[i][j], end = "")
    print()
```
