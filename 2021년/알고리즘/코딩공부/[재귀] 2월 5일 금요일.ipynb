# 하노이  탑: 백준11729번

```python
N = int(input())
cnt = 0
step = []

def move(N,start,to):
    step.append((start,to))
#출발 --> 목적 --> 경유
def hanoi(N,start,to,via):
    global cnt
    cnt += 1
    if N == 1: #N == 1이면 어디든 갈 수 있다.
        move(1,start,to)
    else:
        hanoi(N-1,start,via,to) # N-1개를 경유지로
        move(N,start,to) # N th을 목적지로
        hanoi(N-1,via,to,start) #N-1개를 다시 목적지로 

hanoi(N,1,3,2)

print(cnt)
for i in range(len(step)):
    print(*step[i])
```
