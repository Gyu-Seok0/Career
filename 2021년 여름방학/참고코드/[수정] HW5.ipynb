# Custom DataSet

```python
import os
from skimage import io
 

    
# dataset만들기
class CustomDataset(Dataset):
    def __init__(self, root_dir, transform = None):
        
        self.root_dir = root_dir
        self.transform = transform
        self.x = []
        self.y = []
        
        # 디렉토리 (20개)
        for i in range(20):
            init_path = self.root_dir + "/" + str(i) + "/"

            # 디렉토리 내부(폴더당 2000개씩)
            for j in range(len(os.listdir(init_path))):
                img_path = init_path + os.listdir(init_path)[j] #이미지의 Path
                img = io.imread(img_path) #실제 이미지 읽어오기
                
                # 추가
                self.x.append(img)
                self.y.append([i])
                
    def __len__(self):
        return len(self.x)
    
    def __getitem__(self,idx):
        
        x = self.x[idx]
        y = self.y[idx]
        
        if self.transform:
            x = transforms.ToPILImage()(x) #이미지로 변환
            x = self.transform(x)
#             x = transforms.ToTensor()(x) #다시 Tensor로 변환. (아래 trans에도 있는 구문이기에, 여기서는 주석처리)
            
        x = torch.FloatTensor(x)
        y = torch.FloatTensor(y)

        return x,y
        
        


# transform
trans = transforms.Compose([
#         transforms.Resize((224,224)),
        transforms.ToTensor(),    
])

dataset = CustomDataset(root_dir = "./train", transform = trans )
train_loader = DataLoader(dataset=dataset, batch_size = 100, shuffle = True, drop_last= True)


```

# 정규화를 위한 함수

```python
# function for mean & std
def get_mean_std(loader):
    mu = 0
    mu_square = 0
    std = 0
    for sample in loader:
        X,Y = sample

        mu += torch.mean(X, dim = [0,2,3])
        mu_square += torch.mean(X**2, dim = [0,2,3])
        std += torch.std(X, dim = [0,2,3])

    mu /= len(loader)
    mu_square /= len(loader)
    std = (mu_square - mu**2)**0.5 # E[X**2] - {E[X]}**2
    return mu, std



mu, std = get_mean_std(train_loader)
print(mu, std)
    
```

# Model 이랑 Test 돌려보기

```python
#step2) device
device = "cuda" if torch.cuda.is_available() else "cpu"
torch.manual_seed(777)
if device == "cuda":
    torch.cuda.manual_seed_all(777)

#step3) hyper-parameter
learning_rate = 0.001
batch_size = 100
training_epoch = 10


#step4) dataset & dataloader (위에서 이미 해결)


#step5) 모델링
class CNN(nn.Module):
    def __init__(self,num_classes):
        super(CNN,self).__init__()
        
        # 1*28*28
        self.layer1 = nn.Sequential(
            nn.Conv2d(1,128,3, padding= 1), 
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2),


        )
        
        # 128*14*14
        self.layer2 = nn.Sequential(
            nn.Conv2d(128,128*2,1), 
            nn.BatchNorm2d(128*2),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )
        
        # 256*7*7
        self.layer3 = nn.Sequential( 
            nn.Conv2d(256,256*2,3, groups = 256, padding = 1),
            nn.BatchNorm2d(256*2),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )
        
        # 512*3*3
        self.layer4 = nn.Sequential(
            nn.Conv2d(512,512*2,1),
            nn.BatchNorm2d(512*2),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )
        
        # 1024*1*1
        self.layer5 = nn.Sequential(
            nn.Linear(1024,512),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Dropout(0.5),

            
            nn.Linear(512,num_classes)
        )
        
    def forward(self,x):
        out = self.layer1(x)
        
        out = self.layer2(out)
        
        out = self.layer3(out)
        
        out = self.layer4(out)
        
        out = out.view(out.size(0), -1)
        
        out = self.layer5(out)

        return out
    
    def initalize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode = "fan_out", nonlinearity= "relu")
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.constant_(m.weight,1)
                nn.init.constant_(m.bias,0)
            elif isinstance(m, nn.Linear):
                nn.init.normal_(m.weight, 0, 0.01)
                nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.BatchNorm1d):
                nn.init.constant_(m.weight,1)
                nn.init.constant_(m.bias,0)
        
        

model = CNN(num_classes = 20).to(device)
model.initalize_weights()

test = torch.Tensor(2,1,28,28).to(device)
model(test)

```

```python
#step6) loss & optim
criterion = nn.CrossEntropyLoss().to(device)
optimizer = optim.Adam(model.parameters(), lr = learning_rate)

```

```python
# loss 그래프 그리기
import visdom
vis = visdom.Visdom()

def loss_tracker(loss_plt, loss_value, num):
    vis.line(X = num,
             Y = loss_value,
             win = loss_plt,
             update = "append")
    
loss_plt = vis.line(Y = torch.Tensor(1).zero_(), opts = dict(title = "loss_tracker",
                                                             legend = ["loss"],
                                                             showlegend = True))
```

```python
#step7) train
iteration = len(train_loader)
model.train()

for epoch in range(training_epoch):
    loss = 0
    correct = 0
    for sample in train_loader:
        optimizer.zero_grad()
        
        X,Y = sample
        X = X.to(device)
        Y = Y.to(device)
        Y = torch.squeeze(Y) # 채널 맞추기
        Y = Y.long() # type 맞추기
        
        # forward & backward & optim
        hypothesis = model(X)
        cost = criterion(hypothesis, Y)
        cost.backward()
        optimizer.step()
        
        #calculate
        loss += cost.item()
        correct += (torch.argmax(hypothesis, dim = 1) == Y).float().sum()
    
    loss /= iteration
    acc = correct / (batch_size * iteration)
    print("[Epoch {:04d}, loss = {:.4f}, acc = {:.2f}]%".format(epoch, loss, acc*100))
    
    # 그래프 그리기 및 weight 저장하기.
    loss_tracker(loss_plt, torch.FloatTensor([loss]), torch.FloatTensor([epoch]))
    torch.save(model.state_dict(), "./hw5_weight/%d_%d.pth"%(epoch+1,int(acc*100)))
    
    
```

```python
#step8) test

with torch.no_grad():
    model.eval()
    loss = 0
    correct = 0
    iteration = len(test_loader)

    for sample in test_loader:
        X, Y = sample
        X = X.to(device)
        Y = torch.squeeze(Y)
        Y = Y.long()
        
        hypothesis = model(X)
        cost = criterion(hypothesis, Y)
        
        # calculate
        loss += cost.item()
        correct += (torch.argmax(hypothesis, dim = 1) == Y).float().sum()
    
    
    loss /= iteration
    acc = correct / (iteration*batch_size)
    print("[Test] loss = {:.4f}, acc = {:.2f}%".format(loss, acc*100))
        
        
        
```

# Depthwise_separable_conv에서 
depth_wise 채널 잡는게 조금 신기
--> 인풋채널에 대해서, 몇개의 커널set을 사용할것인지 중요

```python
class depthwise_separable_conv(nn.Module):
    def __init__(self, nin, kernels_per_layer, nout):
        super(depthwise_separable_conv, self).__init__()
        self.depthwise = nn.Conv2d(nin, nin * kernels_per_layer, kernel_size=3, padding=1, groups=nin)
        self.pointwise = nn.Conv2d(nin * kernels_per_layer, nout, kernel_size=1)

    def forward(self, x):
        out = self.depthwise(x)
        out = self.pointwise(out)
        return out
```
