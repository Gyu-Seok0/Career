```python
import matplotlib.pyplot as plt
import csv

f = open('test.csv')
data = csv.reader(f)
next(data)

max_temp = []
for row in data:
    if row[-1] != '' and row[2].split("-")[0] == "Aug":
        max_temp.append(float(row[-1]))
```

```python
plt.hist(max_temp, color= "hotpink", bins=100)
plt.title("highest temperaure in Aug, 1907~2019")
plt.xlabel("temperature")
plt.ylabel("Freq")
plt.show()
```

```python
plt.boxplot(max_temp)
plt.show()
```

```python
plt.pie(max_temp)
plt.show()
```
